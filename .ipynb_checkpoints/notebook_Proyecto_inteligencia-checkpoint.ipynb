```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as io
import math as mt
import random as rand
import os

#################################################################################
# Initial parameters definition
DPI = 0.1
API = 0.1
# Load instance parameters
inst = io.loadmat('gaskell325a.mat')
cx = np.transpose(inst['costumX'])
cx = cx.astype(float)
cy = np.transpose(inst['costumY'])
cy = cy.astype(float)
dx = np.transpose(inst['depotX'])
dx = dx.astype(float)
dy = np.transpose(inst['depotY'])
dy = dy.astype(float)
custom = np.concatenate((cx, cy), axis=1)
depots = np.concatenate((dx, dy), axis=1)


#################################################################################
# Distance Matrix Calculation
# Euclidean distance equation
def eucdist(x1, y1, x2, y2):
    a = pow(abs(x1 - x2), 2)
    b = pow(abs(y1 - y2), 2)
    return np.sqrt(a + b)


control = True
E = []
for i in range(custom.shape[0]):
    for j in range(custom.shape[0]):
        if i != j:
            if control:
                E = np.array([[i, j, custom[i, 0], custom[i, 1], custom[j, 0], custom[j, 1],
                              eucdist(custom[i, 0], custom[j, 0], custom[i, 1], custom[j, 1])]])
                control = False
            else:
                arr = np.array([[i, j, custom[i, 0], custom[i, 1], custom[j, 0], custom[j, 1],
                                eucdist(custom[i, 0], custom[j, 0], custom[i, 1], custom[j, 1])]])
                E = np.concatenate((E, arr), axis=0)
#################################################################################
# Vehicle and Depots' Capacities
k = []
vehCap = inst['vehCap'][0][0].astype(float)
depCap = inst['depCap'][0][0].astype(float)
dems = np.transpose(inst['demands']).astype(float)
depCaps = np.zeros(depots.shape[0]).astype(float)
depCaps[:] = depCap
depCost = inst['openDepCost'].astype(float)
vehCost = 0
#################################################################################
# Phase 1: Depot Ranking
W = np.zeros(depots.shape[0])
P = np.zeros((depots.shape[0], 2))
for i in range(depots.shape[0]):
    dist = 0.0
    for j in range(custom.shape[0]):
        dist = dist + eucdist(depots[i, 0], depots[i, 1], custom[j, 0], custom[j, 1])
    W[i] = dist
    P[i] = np.array([depCaps[i] / (depCost * W[i]), i])
# Order resulting vector P in ascending way
Ps = np.concatenate((P, depots), axis=1)
PSidx = np.argsort(Ps, axis=0)
Ps = Ps[PSidx[:, 0]]


#################################################################################
# Phase 2: Fuzzy Demands
# Function that calculates the fuzzy demand from a deterministic demand


def fuzzyn(d):
    d_1 = d
    d_3 = 3 * d
    d_11 = 1.5 * d_1
    d_31 = 2.5 * d_1
    d_2 = ((d_31 - d_11) * np.random.rand(1)) + d_11
    return [d_1, d_2, d_3]


d2sum = 0
fuzzDems = np.zeros((dems.shape[0], 3))
for i in range(dems.shape[0]):
    [d1, d2, d3] = fuzzyn(dems[i])
    d2sum = d2sum + d2
    fuzzDems[i, :] = [d1, d2, d3]

# number of depots to be opened
depNum = mt.ceil(d2sum / depCap)
ind1 = depots.shape[0] - depNum
ind2 = depots.shape[0]

# Depot selection statement
selDeps = Ps[ind1:, :]
dic = {'API': API, 'DPI': DPI, 'E': E, 'dx': dx, 'dy': dy, 'cx': cx, 'cy': cy, 'k': k, 'P': P, 'Ps': Ps, 'W': W,
       'custom': custom, 'd2sum': d2sum, 'dems': dems, 'depCap': depCap, 'depCaps': depCaps, 'depCost': depCost,
       'depNum': depNum, 'depots': depots, 'fuzzDems': fuzzDems, 'ind1': ind1, 'ind2': ind2, 'inst': inst,
       'selDeps': selDeps, 'vehCap': vehCap, 'vehCost': vehCost}
io.savemat('clusters.mat', mdict=dic)
#################################################################################
# Phase 3: Customer Clustering


def ranf(fuzz):
    r = float(fuzz[0] + ((fuzz[2] - fuzz[0]) * np.random.rand(1)))
    return r


def member(fuzz, dem):  # Membership function
    m = 0
    if dem > fuzz[0] and dem <= fuzz[1]:
        m = (dem - fuzz[0])/(fuzz[1] - fuzz[0])
    elif dem > fuzz[1] and dem < fuzz[2]:
        m = (fuzz[2] - dem)/(fuzz[2] - fuzz[1])
    return m


def fuzzyd(fuzz):
    # Actual customer demand calculation
    m = 1
    r = 0
    dem = 0
    while m > r:
        dem = ranf(fuzz)
        m = member(fuzz, dem)
        r = np.random.rand(1)
    return [fuzz[0], dem, fuzz[2]]


def capsubs(q0, fuzz):
    q = np.copy(q0)
    q[0] = q[0] - fuzz[2]
    q[1] = q[1] - fuzz[1]
    q[2] = q[2] - fuzz[0]
    return q


def seekdist(e, ind):
    selInd = np.zeros([custom.shape[0]-1, 7])
    cont = 0
    for it in range(e.shape[0]):
        if int(e[it, 0]) == ind:
            selInd[cont, :] = e[it, :]
            cont += 1
    sortIdx = np.argsort(selInd[:, -1])
    sortSelInd = selInd[sortIdx, :]
    return sortSelInd


def cred(cap, fzd2):
    cr = 0
    if fzd2[0] > cap[2]:
        cr = 0
    elif fzd2[0] <= cap[2] and fzd2[1] >= cap[1]:
        cr = (cap[2] - fzd2[0])/(2 * (cap[2] - fzd2[0] + fzd2[1] - cap[1]))
    elif fzd2[1] < cap[1] and fzd2[2] >= cap[0]:
        cr = (fzd2[2] - cap[0] - (2 * (fzd2[1] - cap[1])))/(2 * (cap[1] - fzd2[1] + fzd2[2] - cap[0]))
    elif fzd2[2] < cap[0]:
        cr = 1
    return cr


def evalcust(dis, clus, fzd2, vcap, dpi):
    idxvec = []
    for it in range(dis.shape[0]):
        for jt in range(clus.shape[0]):
            if dis[it, 1] == clus[jt]:
                idxvec.append(it)
    if len(idxvec) > 0:
        dis = np.delete(dis, idxvec, 0)
    if dis.shape[0] > 0:
        cust2 = int(dis[0, 1])
        fzd_i = fuzzyd(fzd2[cust2, :])
        cr = cred(vcap, fzd_i)
        if cr >= dpi:
            bin2 = True
            cap = capsubs(vcap, fzd_i)
        else:
            bin2 = False
            cap = vcap
    else:
        bin2 = False
        cust2 = 0
        cap = vcap
    return [bin2, cust2, cap]


def newclus(cter, num):
    vec = np.arange(0, num)
    for it in range(cter.shape[0]):
        for jt in range(int(num)):
            if vec[jt] == cter[it]:
                vec = np.delete(vec, jt, 0)
                break

    if vec.shape[0] > 0:
        rcus = np.random.choice(vec, 1)
    else:
        rcus = 0
    return int(rcus)


# Random choice of the first customer in order to begin clustering
numc = inst['numCostum']
rCus = rand.randint(0, numc - 1)
cterExt = np.array([])
clusCont = 1
os.mkdir('clusters')
while len(cterExt) < numc:
    Q = np.array([vehCap, vehCap, vehCap])  # Vehicle capacity
    fzd = fuzzyd(fuzzDems[rCus, :])
    Q = capsubs(Q, fzd)
    cterInt = np.array([])
    cterInt = np.append(cterInt, rCus)
    cterExt = np.append(cterExt, rCus)
    # Evaluating customers in order of distance
    disl = seekdist(E, rCus)
    deadC = 0
    binVar = True
    while binVar and disl.shape[0] > 0:
        [bin, nCus, Q] = evalcust(disl, cterExt, fuzzDems, Q, DPI)
        if bin:
            cterInt = np.append(cterInt, nCus)
            cterExt = np.append(cterExt, nCus)
            if deadC == numc - cterInt.shape[0]:
                binVar = False
                #Create string for saving the cluster in the matfile
                str1 = 'clusters/cluster'+str(clusCont)
                clusCont += 1
                assigDep = []
                io.savemat(str1, {'cterInt': cterInt, 'cterExt': cterExt, 'assigDep': assigDep})
                rCus = newclus(cterExt, numc)
        else:
            if deadC == numc - cterInt.shape[0]:
                binVar = False
                str1 = 'clusters/cluster'+str(clusCont)
                clusCont += 1
                assigDep = []
                io.savemat(str1, {'cterInt': cterInt, 'cterExt': cterExt, 'assigDep': assigDep})
                rCus = newclus(cterExt, numc)
            deadC += 1
            for i in range(disl.shape[0]):
                if disl[i, 1] == nCus:
                    disl = np.delete(disl, i, 0)
                    break

#################################################################################
# Clusters to depots assignation
def loadmats(sizef, fls):
    mat = []
    for it in range(sizef):
        mi = io.loadmat('./clusters/cluster'+str(it+1)+'.mat')
        mat.append(mi)
    return mat

def gravC(clust, cust):
    cus = cust
    clu = clust['cterInt'][0]
    A = 0
    B = 0
    s = clu.shape[0]
    for i in range(s):
        A += cus[int(clu[i]), 0]
        B += cus[int(clu[i]), 1]
    A = A/s
    B = B/s
    return [A, B]

def sumclusdem(mat, fuzz):
    cter = mat['cterInt'][0]
    dems = fuzz
    s = cter.shape[0]
    dSum = [0, 0, 0]
    for i in range(s):
        dSum += dems[int(cter[i]), :]
    return dSum

def evalclus(mat, depot, api, depcap, ps, cust, cont1):
    dismat = []
    cont = 0
    s = len(mat)
    depC = depcap
    for it in range(s):
        rDep = ps[depot, :]
        gc = gravC(mat[it], cust)
        dis = eucdist(gc[0], gc[1], rDep[2], rDep[3])
        dismat.append([it, gc[0], gc[1], dis])
    dismat.sort(key=lambda x: x[-1])
    bin = True
    it = 0
    fzdcap = [depcap, depcap, depcap]
    while bin and it < s:
        ind = dismat[it][0]
        mat_i = mat[ind]
        #Check if the cluster_i has a depot or not
        if mat_i['assigDep'].shape[0] == 0:
            sDem_i = sumclusdem(mat_i, fuzzDems)
            cr_i = cred(fzdcap, sDem_i)
            if cr_i > api:
                fzdcap -= sDem_i
                strf = 'clusters/cluster'+str(ind+1)+'.mat'
                assigDep = ps[depot, :]
                io.savemat(strf, {'cterInt': mat_i['cterInt'], 'cterExt': mat_i['cterExt'], 'assigDep': assigDep}, appendmat=True)
                it += 1
                cont += 1
            else:
                bin = False
        else:
            it += 1
    return [cont, depot]


files = os.listdir('./clusters')
sf = len(files)
sd = Ps.shape[0]
mats = loadmats(sf, files)
#c1 = mats[0]['cterInt'][0]
i = 1
cont = 0
deps = []
while cont < sf:
    v = sd-i
    [inds, depot] = evalclus(mats, v, API, depCap, Ps, custom, cont)
    cont += inds
    dep = Ps[depot, :]
    deps.append(dep)
    mats = loadmats(sf, files)
    i += 1

#################################################################################
# Genetic Algorithm

def initial(cust):
    s = cust.shape[0]
    if s <= 3:
        perm = np.zeros([10, s])
        for i in range(10):
            j = np.random.permutation(cust)
            perm[i, :] = j
    elif s > 3 and s <= 6:
        perm = np.zeros([100, s])
        for i in range(100):
            j = np.random.permutation(cust)
            perm[i, :] = j
    elif s > 6 and s <= 9:
        perm = np.zeros([1000, s])
        for i in range(1000):
            j = np.random.permutation(cust)
            perm[i, :] = j
    elif s > 9:
        perm = np.zeros([10000, s])
        for i in range(10000):
            j = np.random.permutation(cust)
            perm[i, :] = j
    return perm

def evalcus2(cust, fzd, vcap, dpi):
    cust2 = None
    if cust.shape[0] > 0:
        cust2 = int(cust[0])
        fzd1 = fzd[cust2, :]
        fzd2 = fuzzyd(fzd1)
        cr = cred(vcap, fzd2)
        if cr >= dpi:
            bin = True
            cap = capsubs(vcap, fzd2)
        else:
            bin = False
            cap = vcap
    else:
        bin = False
        cap = vcap
    return [bin, cap, cust2]

def evalsol(pob_i, cust, dems, dep, vcap, vcapini, dcap, dpi):
    clus1 = np.copy(pob_i)
    dist = 0
    dist2 = 0
    vfzcap = np.copy(vcap)
    xc = dep[2]
    yc = dep[3]
    s = clus1.shape[0]
    clus2 = np.array([])
    i = 0
    while i < s:
        [bin, cap, cust1] = evalcus2(clus1, dems, vfzcap, dpi)
        vfzcap = cap
        xd = cust[cust1, 0]
        yd = cust[cust1, 1]
        if bin:
            d = eucdist(xc, yc, xd, yd)
            dist += d
            clus2 = np.append(clus2, cust1)
            clus1 = np.delete(clus1, 0)
            cust2 = cust[int(clus2[-1]), :]
            xc = cust2[0]
            yc = cust2[1]
            i += 1
        else:
            xd = dep[2]
            yd = dep[3]
            dist2 += eucdist(xc, yc, xd, yd)
            dcap -= (vcapini - vfzcap[1])
            vfzcap = vcap
    sumd = dist + dist2
    return [dist, dist2, sumd]

def evaluation(inipob, cust, dems, dep, vehc, depc, dpi):
    vCap = np.array([vehc, vehc, vehc])
    vCapIni = float(np.copy(vehc))
    s = inipob.shape[0]
    vecDist = np.zeros([s, 3])
    for i in range(s):
        [dist, dist2, sumd] = evalsol(inipob[i, :], cust, dems, dep, vCap, vCapIni, depc, dpi)
        vecDist[i, :] = [dist, dist2, sumd]
    evaluated = np.concatenate((inipob, vecDist), axis=1)
    return evaluated


def selection(ipob, prob):
    pob = np.copy(ipob)
    s1 = pob.shape[0]
    s2 = pob.shape[1]
    np.random.shuffle(pob)
    h = int(s1/2)
    selected = np.zeros([h, s2])
    half1 = pob[:h, :]
    half2 = pob[h:, :]
    for i in range(h):
        if half1[i, -1] < half2[i, -1] and np.random.rand(1) > prob:
            selected[i, :] = half1[i, :]
        else:
            selected[i, :] = half2[i, :]
    return selected


def crossover(sel):
    s1 = sel.shape[0]
    s2 = sel.shape[1]
    h = int(s1/2)
    parents1 = sel[:h, :]
    parents2 = sel[h:, :]
    child1 = np.zeros([h, s2])
    child2 = np.zeros([h, s2])
    for i in range(h):
        cont1 = 0
        cont2 = 0
        bin = False
        p1 = parents1[i, :]
        p2 = parents2[i, :]
        h1 = np.zeros([1, s2])
        h2 = np.zeros([1, s2])
        for j in range(s2):
            if p1[j] == p2[j]:
                cont1 += 1
            else:
                cont2 = cont1
                cont1 = 0
                if cont2 >= 2:
                    bin = True
                    h1 = np.concatenate((p2[j:], p1[j - cont2:j], p2[:j - cont2]))
                    h2 = np.concatenate((p1[j:], p2[j - cont2:j], p1[:j - cont2]))
                    cont2 = 0
        if not bin:
            h1 = p1
            h2 = p2
        child1[i, :] = h1
        child2[i, :] = h2
    crossed = np.concatenate((parents1, parents2, child1, child2))
    np.random.shuffle(crossed)
    return crossed


def mutation(cross, prob):
    s1 = cross.shape[0]
    s2 = cross.shape[1]
    for i in range(s1):
        if np.random.rand(1) > prob:
            r1 = np.random.randint(1, s2 - 1)
            while r1 == s2-1:
                r1 = np.random.randint(1, s2 - 1)
            r2 = np.random.randint(1, s2 - 1)
            while r2 < r1:
                r2 = np.random.randint(1, s2 - 1)
            c = np.concatenate((cross[i, r1:r2], cross[i, :r1], cross[i, r2:]))
            cross[i, :] = c
    return cross


def genetic(mat, iterat, probs, probm):
    clus = mat['cterInt'][0]
    dep = mat['assigDep'][0]
    vehC = float(np.copy(vehCap))
    depC = float(np.copy(depCap))
    dems = np.copy(fuzzDems)
    dpi = float(np.copy(DPI))
    cust = np.copy(custom)
    #Initial Pob
    iniPob = initial(clus)
    #Evaluation
    evaluated = evaluation(iniPob, cust, dems, dep, vehC, depC, dpi)
    s2 = evaluated.shape[1]
    mins = np.zeros([iterat, s2])
    infi = np.zeros([1, s2])
    sidx = np.argsort(evaluated[:, -1])[0]
    infi[0, :] = evaluated[sidx, :]
    for i in range(iterat):
        print('Iteration: '+str(i))
        selected = selection(evaluated, probs)
        selected = np.delete(selected, -1, axis=1)
        selected = np.delete(selected, -1, axis=1)
        selected = np.delete(selected, -1, axis=1)
        crossed = crossover(selected)
        mutated = mutation(crossed, probm)
        evaluated = evaluation(mutated, cust, dems, dep, vehC, depC, dpi)
        idxmin = np.argsort(evaluated[:, -1])[0]
        mins[i, :] = evaluated[idxmin, :]
        if mins[i, -1] <= infi[-1, -1]:
            infi = np.append(infi, np.array([mins[i, :]]), axis=0)
        else:
            infi = np.append(infi, np.array([infi[-1, :]]), axis=0)
    return [evaluated, mins, infi, dep]

#################################################################################
#Test genetic algorithm
for i in range(len(mats)):
    probs = 0.5
    probm = 0.5
    iterat = 50
    [solutions, mins, infi, dep_i] = genetic(mats[i], iterat, probs, probm)
    x = np.arange(iterat)
    ymins = mins[:, -1]
    yinfi = infi[:-1, -1]
    min_cords_idx = infi[-1, :-3]
    plt.figure(1)
    plt.subplot(111)
    plt.plot(x, ymins, 'r-o', label='Mins by Iter')
    plt.plot(x, yinfi, 'b-o', label='Infimun Vals')
    plt.title('Genetic Results:')
    plt.xlabel('Iterations')
    plt.ylabel('Distance Units')
    plt.legend()
    plt.show()
    print('Resulting route for cluster : ' + str(i))
    print(min_cords_idx)
```
